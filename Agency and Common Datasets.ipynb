```python
fields = ['EID', 'LOC_COM' ,'ZIP','latitude','longitude']
df = pd.read_csv('/Users/sarap/Desktop/Data_Science/Capstone/Common_1.csv', usecols=fields)
print ("Open CSV")
```

```python
df.head()
```

```python
engine = create_engine('postgresql+psycopg2://postgres:xxx@localhost:5432/Capstone')
df.to_sql('common_1', engine,if_exists='replace')
```

```python
#extract everything from the agency_1 table and common_1 into a dataframe
agency_1=pd_sql.read_sql('select * from agency_1', con=engine,index_col='index').reset_index()
common_1=pd_sql.read_sql('select * from common_1', con=engine,index_col='index').reset_index()
```

```python
#An outer merge, outer keeps all key values in both data frames
outer_merge=pd.merge(agency_1,
                   common_1,
                   how="outer",
                   on="EID")

outer_merge
```
